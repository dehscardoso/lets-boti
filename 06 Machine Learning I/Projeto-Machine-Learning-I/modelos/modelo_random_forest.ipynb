### Seleção de Features

Para aplicação do modelo foram selecionadas três variáveis que apresentaram menor correlação com a variável target, sendo elas: `EXT_SOURCE_1, EXT_SOURCE_2, EXT_SOURCE_3`, em que cada uma representa o score normalizado a partir de uma fonte de dados externa. 

E, posteriormente, as duas variáveis que apresentaram a maior correlação positiva, sendo elas: `DAYS_BIRTH` e `REGION_RATING_CLIENT_W_CITY`, referindo-se a idade do cliente em dias no momento do pedido e a classificação da região na qual o cliente vive, respectivamente.

### Aplicação do Modelo no Dataset `application_train` com Balanceamento

```python
from imblearn.over_sampling import SMOTE

# preenchendo valores nan com 0
df = df.fillna(0)

# separando dados
X = df.drop(['TARGET'], axis = 1).values
y = df[['TARGET']].values
```

```python
# separação dos dados
x_treino, x_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.2, random_state = 4)

# tratando o desbalanceamento das classes da variável target
smote = SMOTE()
X_smote, y_smote = smote.fit_resample(x_treino, y_treino)
    
modelo = RandomForestClassifier(n_estimators = 100, random_state = 4)
modelo.fit(X_smote, y_smote)

y_teste_pred = modelo.predict_proba(x_teste)[:, 1]
y_pred = modelo.predict(x_teste)
```

### Métricas de Avalição do Modelo

```python
print('Score de treino: {:.4f}'.format(modelo.score(X_smote, y_smote)))
print('Score de teste: {:.4f}'.format(modelo.score(x_teste, y_teste)))
```

```python
print('Acurácia:', accuracy_score(y_teste, y_pred))
print('-------------')
print(classification_report(y_teste, y_pred))
```

```python
# plot curva ROC
bc = BinaryClassification(y_teste, y_teste_pred, labels = [0, 1])
plt.figure(figsize = (5,5))
bc.plot_roc_curve()
plt.show()
```

```python
# matriz de confusão
cm = confusion_matrix(y_teste, y_pred)
plot_confusion_matrix(cm)
plt.show()
```
