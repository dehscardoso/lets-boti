## Funções

```python
## Função para remover outliers

def outlier_detect(df):
    
    for i in df.describe().columns:
        Q1  = df.describe().at['25%',i]
        Q3  = df.describe().at['75%',i]
        
        IQR = Q3 - Q1
        LTV = Q1 - 1.5 * IQR
        UTV = Q3 + 1.5 * IQR
        
        x = np.array(df[i])
        p = []
        for j in x:
            if j < LTV or j > UTV:
                p.append(df[i].median())
            else:
                p.append(j)
        df[i] = p
    return df
```
