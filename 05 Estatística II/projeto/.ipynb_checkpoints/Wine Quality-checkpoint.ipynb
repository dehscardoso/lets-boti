## Importando Bibliotecas

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# importando dados
red_wine_df   = pd.read_csv('winequality-red.csv', sep=";")
white_wine_df = pd.read_csv('winequality-white.csv', sep=";")
```

```python
# 5 primeiras observações do dataset de vinhos tintos
red_wine_df.head()
```

```python
# 5 últimas observações do dataset de vinhos tintos
red_wine_df.tail()
```

```python
# 5 primeiras observações do dataset de vinhos brancos
white_wine_df.head()
```

```python
# 5 últimas observações do dataset de vinhos brancos
white_wine_df.tail()
```

## Juntando Datasets

```python
# Criando a variável type para identificar o tipo do vinho 
# de cada observação antes de fazer a concatenação dos dados
red_wine_df['type']   = 'red'
white_wine_df['type'] = 'white'
```

```python
# Concatenando os dados 
wine_df = pd.concat([red_wine_df, white_wine_df])
wine_df.reset_index(drop=True, inplace=True)
```

```python
# 5 primeiras observações do dataset
wine_df.head()
```

```python
# 5 últimas observações do dataset
wine_df.tail()
```

```python
# informações sobre o dataset
wine_df.info()
```

## Tratamento das Variáveis

```python
# Percentual de dados faltantes 
wine_df.isnull().mean().round(4)*100
```

```python
# Medidas resumo das variáveis quantitativas
wine_df.drop(columns=['type']).describe(percentiles = [.25, .5, .75, .95, .99]).round(2)
```

### Avaliando Frequências das Variáveis Qualitativas

```python
sns.set(font_scale=1.2)
fig, eixo = plt.subplots(3,4,figsize=(20,12))

cont = 0
for linha in range(3):
    for coluna in range(4): 
        sns.histplot(data=wine_df, x=wine_df.columns[cont], ax=eixo[linha][coluna])
        cont += 1
```

```python
sns.set(font_scale=1.2)
fig, eixo = plt.subplots(3,4,figsize=(20,12))

cont = 0
for linha in range(3):
    for coluna in range(4): 
        sns.boxplot(data=wine_df, x=wine_df.columns[cont], ax=eixo[linha][coluna], orient='h')
        cont += 1
```

```python
sns.set_context(font_scale=1.9)
plt.figure(figsize=(12,8))
sns.heatmap(
    wine_df.corr(), #Fonte dos dados, nossa matriz de correlação
    annot=True, #Anotar os valores das correlações nas caixinhas coloridas
    fmt='.2f', # Formato float de 2 dígitos, para arredondar os números
    cmap='RdBu', # Mapa de cores do vermelho (menor) para o azul (maior)
    mask=np.triu( # Faz uma matriz triangular
        np.ones_like(wine_df.corr()) # usando como base uma matriz de '1's, do mesmo tamanho que a matriz de correlação
    ) # A máscara vai então 'apagar' a diagonal superior da matriz
);
```

## Modelagem dos Dados

```python
# Construindo uma target binária
wine_df['approved'] = (wine_df['quality'] >= 6).astype('int')
```

```python
# Aprovado = 1
# Não aprovado = 0
wine_df.approved.value_counts()
```

```python
# Amostrando
dados_teste  = wine_df.dropna().sample(frac=0.2,random_state=42).drop(['quality'],axis=1)
dados_treino = wine_df.drop(dados_teste.index).drop(['quality'],axis=1)
```
