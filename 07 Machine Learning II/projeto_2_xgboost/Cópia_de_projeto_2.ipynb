```python
import pandas as pd
import numpy as np

from xgboost import XGBClassifier

import seaborn as sns

from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix

from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import OneHotEncoder


import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("/content/drive/MyDrive/lets code/modulo_machine_learning_1/credito-imoveis/application_train.csv", encoding='latin1')
df_train=df_train.drop_duplicates()
df_train.head()
```

```python
df_train_model=df_train[['NAME_CONTRACT_TYPE',
                          'CODE_GENDER',
                          'FLAG_OWN_CAR',
                          'CNT_CHILDREN',
                          'AMT_INCOME_TOTAL',
                          'DAYS_BIRTH',
                          'DAYS_EMPLOYED',
                          'REGION_RATING_CLIENT',
                          'REGION_RATING_CLIENT_W_CITY',
                          'ORGANIZATION_TYPE',
                          'EXT_SOURCE_1',
                          'EXT_SOURCE_2',
                          'EXT_SOURCE_3',
                          'YEARS_BEGINEXPLUATATION_AVG',
                          'FLAG_DOCUMENT_2',
                         'TARGET']]

df_train_model
```

```python
df_train_model.info()
```

```python
# SeparaÃ§Ã£o dos dados em atributos (X) e variÃ¡vel resposta (y)
X = df_train_model.drop('TARGET', axis = 1)
y = df_train_model['TARGET']

# SeparaÃ§Ã£o da base em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X,                 # atributos
                                                    y,                 # VariÃ¡vel resposta
                                                    test_size = 0.3,   # proporÃ§Ã£o base de treino
                                                    random_state = 42, # semente aleatÃ³rio
                                                    stratify = y)      # mantÃ©m a proporÃ§Ã£o das classes
```

```python
pipe_numeric_mode= Pipeline(steps = [
    ('missing_mode', SimpleImputer(strategy = 'most_frequent')),
    ('standartscale', StandardScaler())
])

pipe_numeric_median= Pipeline(steps = [
    ('missing_median', SimpleImputer(strategy = 'median')),
    ('standartscale', StandardScaler())
])

pipe_categ = Pipeline(steps = [
    ('missing_const', SimpleImputer(strategy = 'constant')),
    ('OrdinalEncoder', OrdinalEncoder(handle_unknown='use_encoded_value',
                                      unknown_value=-1)),
])
```

```python
pipe_escalona = ColumnTransformer(transformers = [
    ('pipe_numeric_mode', pipe_numeric_mode, ['CNT_CHILDREN', 'REGION_RATING_CLIENT', 'REGION_RATING_CLIENT_W_CITY','FLAG_DOCUMENT_2']),
    ('pipe_numeric_median', pipe_numeric_median, ['AMT_INCOME_TOTAL',	'DAYS_BIRTH',	'DAYS_EMPLOYED','EXT_SOURCE_1',	'EXT_SOURCE_2', 'EXT_SOURCE_3',	'YEARS_BEGINEXPLUATATION_AVG']),
    ('pipe_cat', pipe_categ, ['NAME_CONTRACT_TYPE',	'CODE_GENDER',	'FLAG_OWN_CAR', 'ORGANIZATION_TYPE'])
])
```

```python
parametros = {
    'modelo__num_leaves': [5, 10, 15],
    'modelo__max_depth': [5, 8, 10],
    'modelo__n_estimators': [500],
    'modelo__learning_rate': [0.01, 0.05],
}

model = XGBClassifier(random_state = 42, 
                      eval_metric="auc", 
                      scale_pos_weight=10)


pipeline_completa = Pipeline(steps = [
    
    ('preproc', pipe_escalona),
    ('modelo', model)

    ])

gscv = GridSearchCV(
    estimator = pipeline_completa,
    param_grid = parametros,
    scoring = 'roc_auc',
    refit = True,
    cv = 3
)
```

```python
gscv.fit(X_train, y_train)
```

```python
gscv.best_params_
```

```python
y_pred = gscv.predict(X_test)
y_pred_prob = gscv.predict_proba(X_test)[:,1]


# calculo da acuracia
print("AcurÃ¡cia",accuracy_score(y_test, y_pred))
print("-------------")
print(classification_report(y_test, y_pred))
print("------")
print("Area curva ROC:", roc_auc_score(y_test, y_pred_prob))
```
