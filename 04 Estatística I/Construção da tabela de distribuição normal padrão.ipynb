```python
import numpy as np
```

```python
#Gaussiana (distribuição normal padrão)
def f(z):
    return (1/np.sqrt(2*np.pi)) * np.exp(-z**2/2)
```

```python
#Método 1/3 de Simpson
def simpson(xo,zo=0):
    n = 100
    soma = 0
    h = xo/n 
    
    soma = f(0)+f(xo)
    for i in range(1,n):
        if i%2 == 0:
            soma += 2 * f(h*i)
        else:
            soma += 4 * f(h*i)
    return zo+(h/3)*soma
    
```

```python
#Construção da tabela de distribuição normal padrão
#zo = 0  indica que é a tabela que calcula área no intervalo 0 <= z <= zc
#zo = 0.5 indica que é a tabela que calcula área no intervalo -infinito <= z <= zc
#dec é número de casas decimais

def tabela(zo=0, dec=4):
    
    linha = '        '
    for i in range (10):
        #linha += str(i) + '        '
        linha += str(i) + (dec+2)*' '
    print(linha)
    print(4*' '+((10*dec+30)*'-'))
    
    for i in range(40):
        print(f'{i*0.1:1.1f}|',end=' ')
        for j in range(10):
            print(f'{(simpson(i*0.1+j*0.01,zo)):1.{dec}f}',end=' ')
        print()

tabela()
```
