P(B) = 14/32 --> P(~ B) = 18/32
P(A e B) = 4/32 -->?

P(A) = 0.4 (futebol)
P(B) = 0.3 (natação)
P(A e B) = 0.2 * 0.4 = 0.08 (ambos)
P(A ou B) = P(A) + P(B) - P(A e B)
          = 0.4 + 0.3 - 0.08 = 0.62
P(A ou B) = 0.62

P(~ (A ou B)) = 1 - P(A ou B) = 1 - 0.62 = 0.38

```python
from random import randint
import pandas as pd

cont = 0
tamanho_da_turma = 45
numero_de_turmas = 10
for _ in range(numero_de_turmas):
    aniversarios = []
    for _ in range(tamanho_da_turma):
        aniversarios.append(randint(1,365))
    aniversarios = pd.Series(aniversarios)
    if len(aniversarios.unique()) < tamanho_da_turma:
        #print('Mesmo dia!!',j)
        cont+=1

print("Total de vezes com dois aniversários no mesmo dia:", cont)
```
